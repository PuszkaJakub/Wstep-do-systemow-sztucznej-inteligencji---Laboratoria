```python
# Wygeneruj tablicę o rozmiarze 10000 zawierającą losowe wartości z rozkładu normalnego wartości oczekiwanej  μ=1  
# i odchyleniu standardowym  σ=3  i wyrysuj histogram dla tych danych.
```

```python
import matplotlib.pyplot as plt
import numpy as np

wart_oczekiwana = 1
odchyl_standard = 3

a = np.array(wart_oczekiwana + odchyl_standard * np.random.randn(10000))     
plt.hist(a)
```
